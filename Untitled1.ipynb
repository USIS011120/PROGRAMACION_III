```python
import pandas as pd
import matplotlib.pyplot as plt

TarjeCredito = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",")

getSexo = pd.crosstab(index=TarjeCredito["Gender"], columns="Frecuencia")
print(getSexo)

pieSexo = getSexo.loc[getSexo.index]["Frecuencia"]
plt.pie(pieSexo, labels=pieSexo.index)
plt.pie(pieSexo, autopct='%1.1f%%', shadow=True, startangle=120)
plt.xlabel("TARJETAS DE CREDITO POR SEXO")
plt.rcParams['figure.facecolor'] = 'mistyrose'
plt.show

getSexo.to_excel("/content/tarjetas_por_sexo.xlsx")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

tarejetas_credito = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",")

tipo_cuidad = tarejetas_credito.groupby(["City", "Card Type"]).sum()
print(tipo_cuidad)

pietipo_cuidad = tipo_cuidad.loc[tipo_cuidad[:10].index]#limitacion de la cantiadad de datos 
plt.rcParams['figure.facecolor'] = 'skyblue'
colores = ['red','green','blue','yellow','cyan','coral','plum','silver','magenta','orange']
plt.pie(pietipo_cuidad, labels=pietipo_cuidad.index,colors = colores)
plt.title("Tipo de tarjetas de cr√©dito por ciudades")
plt.show

tipo_cuidad.to_excel("Total_dolares_tarjetas_cuidad.xlsx")

```
